Published-value comparison note:

- The code cell below sketches a compact visual comparison for the main timing ranges.


```python
import matplotlib.pyplot as plt

labels = ["Tegmark", "Hagan reply", "Actin gel", "Repo timing stats"]
published_low = [1e-20, 1e-5, 1e-2, 1.947002e-02]
published_high = [1e-13, 1e-4, 1e-1, 3.210064e-02]
repo_value = [3.162278e-17, 3.162278e-05, 3.162278e-02, 2.500000e-02]

fig, ax = plt.subplots(figsize=(8, 3))
y = range(len(labels))
for i, (low, high, repo) in enumerate(zip(published_low, published_high, repo_value)):
    ax.hlines(i, low, high, color="0.7", linewidth=6)
    ax.scatter(repo, i, color="C0", s=30)
ax.set_xscale("log")
ax.set_yticks(list(y))
ax.set_yticklabels(labels)
ax.set_xlabel("seconds")
ax.set_title("Published timing ranges vs frozen repo values")
fig.tight_layout()
fig
```
