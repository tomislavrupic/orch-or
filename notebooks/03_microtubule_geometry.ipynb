# Microtubule Geometry and `E_G` Sensitivity

Executable HAOS-OR notebook for microtubule geometry, mass distribution, and `E_G` model comparison. This remains a diagnostic scaffold, not a biological proof.

Primary-structure candidates to trace next:

- Nogales, Wolf, Downing 1998, *Structure of the alpha beta tubulin dimer by electron crystallography*.
- Nogales, Whittaker, Milligan, Downing 1999, *High-resolution model of the microtubule*.
- Lowe, Li, Downing, Nogales 2001, *Refined structure of alpha beta-tubulin at 3.5 A resolution*.

```python
from orch_or.geometry import DEFAULT_GEOMETRY, collapse_time_for_domain

geometry = DEFAULT_GEOMETRY
rows = []
for eg_model in ["gaussian", "uniform_cylinder", "quadrature"]:
    eg_j, tau_s = collapse_time_for_domain(
        geometry=geometry,
        coherent_dimers=13,
        separation_m=1.0e-9,
        smearing_radius_m=1.0e-9,
        eg_model=eg_model,
    )
    rows.append((eg_model, eg_j, tau_s))

rows
```
