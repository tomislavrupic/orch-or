# Literature Comparison

A small notebook for comparing source-traced literature rows against the frozen Orch-OR diagnostics. It stays stdlib-first and uses matplotlib only for plots.

```python
from pathlib import Path
import csv
import json
import sys

import matplotlib.pyplot as plt

ROOT = Path.cwd()
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

from orch_or.data import registry

registry.validate()
primary = registry.load_primary_structure()
summary = json.loads((ROOT / 'examples' / 'output' / 'orch_or_summary.json').read_text())
len(primary), summary['reduction_rows']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

names = [row['parameter'] for row in primary]
values = [float(row['value']) for row in primary]
axes[0].barh(names, values, color='#4c78a8')
axes[0].set_title('Primary Structure Rows')
axes[0].set_xlabel('Value')

summary_keys = ['reduction_rows', 'reduction_sweep_rows', 'reduction_distinction_rows']
summary_values = [summary[key] for key in summary_keys]
axes[1].bar(summary_keys, summary_values, color='#f58518')
axes[1].set_title('Frozen Reduction Artifacts')
axes[1].tick_params(axis='x', rotation=20)

fig.suptitle('Literature and Model Comparison')
fig.tight_layout()
plt.show()
```

Interpretation boundary: these plots show traceable rows and frozen outputs side by side. They do not establish Orch-OR as a biological explanation.
